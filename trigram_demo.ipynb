```python
from trigram_model.tgn import TrigramNetwork
```

```python
network = TrigramNetwork('data/hungarian_cities.csv')
```

```python
network.rate = 20
```

```python
for i in range(500):
    print(f'{i:5} ', end='')
    network.train()
```

```python
# Predict a city name
for i in range(20):
    letters:list[str] = ['<b>', '<b>']
    while True:
        sample = network.sample (
            network.tokenizer.stoi[letters[-2]],
            network.tokenizer.stoi[letters[-1]]
        )
        letters.append(network.tokenizer.itos[int(sample)])
        if (letters[-1] == '<b>'):
            break
    
    s = ''.join(letters).replace('<b>', '')
    if s != '':
        print(s)
```
